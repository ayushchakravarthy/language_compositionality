```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

Transformer Results

SCAN simple split

```python
t_simple_res = []
for i in range(10):
    with open(f'../results/transformer/scan/simple/train_defaults_simple{i}.json', 'r') as f:
        t_simple_res.append(json.load(f))
```

```python
for i in range(10):
    plt.plot(t_simple_res[i]['loss_data'])
```

```python
for i in range(10):
    plt.plot(t_simple_res[i]['train_accs'])
```

```python
for i in range(10):
    plt.plot(t_simple_res[i]['dev_accs'])
```

```python
for i in range(10):
    plt.plot(t_simple_res[i]['test_accs'])
```

```python
t_simple_best = []
for i in range(10):
    t_simple_best.append(max(t_simple_res[i]['test_accs']))
print(np.mean(t_simple_best), '+-', np.std(t_simple_best))
```
